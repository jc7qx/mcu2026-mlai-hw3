# NumPy 陣列 (ndarray) 與 Python 列表 (list) 的差異
- **性能與速度**：NumPy 在數值計算上比 Python 列表快上數倍至數十倍，底層由 C 語言實作，且陣列在記憶體中是連續儲存
- **記憶體佔用**：NumPy 陣列結構更緊湊，佔用的記憶體空間遠少於相同數據量的 Python 列表
- **數據類型**：NumPy 陣列中的所有元素必須是相同類型-同質性（如全部為 `float64`），而 Python 列表可以同時包含不同類型的數據（如 `[1, "a", 3.14]`）
- **數學與矩陣運算**：NumPy 支援廣播 (Broadcasting) 和強大的向量化操作（如 a + b會執行元素對應加法），而列表相加會導致列表長度變長。
- **應用場景**：NumPy 適用於機器學習、資料分析與科學計算等大數據處理；Python 列表適用於儲存少量、異質、需要經常增加或刪除元素的情況

# 線性代數

線性代數與機率統計是機器學習應用的基。**向量(vector)** 運作與 **矩陣(matrix)** 運作是線性代數運算基礎單元。
> 向量同時具有數值和方向，且滿足平行四邊形法則的幾何對象。向量是數學、物理學和工程科學等多個自然科學領域中的基本概念。向量之間可以如數字一樣進行運算。常見的向量運算有：加法、減法、數與向量之間的乘法（數量積)以及向量與向量之間的乘法（外積），但向量的除法沒有定義

> 矩陣（Matrix）是線性代數的基礎，指將實數或複數以長方形陣列（矩形陣列）方式排列的數字集合。它由$m$行$n$列的元素組成，常用於表示線性映射、解線性方程組、圖像處理及機器學習等領域，是數學與工程計算中用於高效處理多維數據的重要工具。

$$A = \begin{bmatrix}
a_{1,1} &a_{1,2} &a_{1,3} \\
a_{2,1} &a_{2,2} &a_{2,3}
\end{bmatrix}$$
- $A$是一個$2x3$($2$列$3$行)的矩陣
- $a_{i,j}$是 $A$ 的元素

矩陣可以表示線性轉換，如旋轉、縮放和投影。支援加法、減法、數乘（標量乘法）和轉置。矩陣乘法不僅是元素的對應相乘，而是行與列的內積組合，代表線性映射的複合。矩陣可以應用於三維動畫製作、計算機圖形學、Data Matrix二維條碼。深度學習中數據通常以張量（多維矩陣）形式儲存和處理。

####向量內積
$\vec{a}\cdot\vec{b}=|\vec{a}||\vec{b}|cos(\angle{ab})=a_x\cdot b_x + a_y \cdot b_y$

```python
import numpy as np
a= np.array([3, 4])
b = np.array([5, 1])
res = np.dot(a, b)
print("a,b向量內積 = ", res)
```

$cos(\angle{ab})=\frac{\vec{a} \cdot \vec{b}}{|\vec{a}|*|\vec{b}|}$ $⇒$ $\angle{ab}=cos^{-1}\frac{\vec{a} \cdot \vec{b}}{|\vec{a}|*|\vec{b}|}$

```python
ang_ab = np.arccos(res/(np.linalg.norm(a)*np.linalg.norm(b)))
```

將$\angle{ab}$由徑度量轉為角度:
2$\pi=360^{\circ} ⇒ 1徑度=\frac{360}{2\pi}=\frac{180}{\pi}$

```python
print("a,b向量夾角 = ", ang_ab*180.0/np.pi)
```

反矩陣（Inverse Matrix，記作$A^{-1}$)，$A$是一個方陣。若存在一個方陣$B$使得 $AB=BA=I$，$I$為單位矩陣，$B$稱為$A$的反矩陣。因此
- $AA^{-1}=A^{-1}A=I$
- $A$必須是方陣，$A$的行列式值不為$0$

反矩陣特性
- $(A^{-1})^{-1} = A$
- $(AB)^{-1}=B^{-1}A^{-1}$
- $(A^{T})^{-1}=(A^{-1})^T$

反矩陣可用於解線性方程式:

$Ax=b$，若$A$可逆，$x=A^{-1}b$

```python
#矩陣x的行列式值
if (np.linalg.det(x) != 0.0):
  print("有反矩陣")
else:
  print("無反矩陣")
```

```python
#反矩陣
np.linalg.inv(x)
```

#解三元一次聯立方程

\begin{eqnarray}
  \left\{
    \begin{array}{l}
      x + y + z = 9 \\
      2x + 3y + 4z = 29 \\
      3x + 2y -z = 8
    \end{array}
  \right.
\end{eqnarray}

```python
#建立A，b矩陣
A = np.array([[1,1,1],[2,3,4],[3,2,-1]])
b = np.array([9,29,8])
```

```python
A
```

```python
#計算A的行列式值(det)，det <> 0 可生成反矩陣
np.linalg.det(A) != 0
```

```python
#建立A反矩陣
np.linalg.inv(A)
```

```python
#解=A反矩陣與b乘積
np.dot(np.linalg.inv(A),b)
```

```python
#numpy提供解聯立方程式solver
x = np.linalg.solve(A,b)
print(x)
```
